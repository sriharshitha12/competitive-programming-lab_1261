```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end_of_word

    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True

contacts = ["Anil", "Anitha", "Anirudh", "Bala", "Balaji"]

trie = Trie()

# Insert contacts
for name in contacts:
    trie.insert(name)

# Operations
print('Contact "Anil" found →', trie.search("Anil"))
print('Contact "Anand" found →', trie.search("Anand"))
print('Prefix "Ani" exists →', trie.starts_with("Ani"))
print('Prefix "Bal" exists →', trie.starts_with("Bal"))
```
