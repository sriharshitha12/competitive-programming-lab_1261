```python


import bisect

# Generate all subset sums of a list
def subset_sums(arr):
    sums = []
    n = len(arr)
    for mask in range(1 << n):
        total = 0
        for i in range(n):
            if mask & (1 << i):
                total += arr[i]
        sums.append(total)
    return sums


def closest_subset_sum(N, S, A):

    # Split array into two halves
    mid = N // 2
    left = A[:mid]
    right = A[mid:]

    # Get subset sums
    left_sums = subset_sums(left)
    right_sums = subset_sums(right)

    # Sort right side for binary search
    right_sums.sort()

    min_diff = float('inf')

    # Find closest sum
    for l in left_sums:
        target = S - l

        idx = bisect.bisect_left(right_sums, target)

        if idx < len(right_sums):
            min_diff = min(min_diff, abs(l + right_sums[idx] - S))

        if idx > 0:
            min_diff = min(min_diff, abs(l + right_sums[idx - 1] - S))

    return min_diff


# Input
T = int(input())

for _ in range(T):
    N, S = map(int, input().split())
    A = list(map(int, input().split()))

    print(closest_subset_sum(N, S, A))
```
