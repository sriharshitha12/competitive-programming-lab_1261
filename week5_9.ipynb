```python
class DSU:
    def __init__(self, n):
        # Parent array
        self.parent = [i for i in range(n + 1)]
        # Rank array
        self.rank = [0] * (n + 1)

    # Find with Path Compression
    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    # Union by Rank
    def union(self, a, b):
        rootA = self.find(a)
        rootB = self.find(b)

        if rootA != rootB:

            if self.rank[rootA] < self.rank[rootB]:
                self.parent[rootA] = rootB

            elif self.rank[rootA] > self.rank[rootB]:
                self.parent[rootB] = rootA

            else:
                self.parent[rootB] = rootA
                self.rank[rootA] += 1


# ============ MAIN ============

# Read number of intersections
while True:
    try:
        n = int(input("Enter number of intersections (a single integer): "))
        break
    except ValueError:
        print("Invalid input. Please enter a single integer.")

# Create DSU
dsu = DSU(n)

# Read number of roads
while True:
    try:
        m = int(input("Enter number of roads (a single integer): "))
        break
    except ValueError:
        print("Invalid input. Please enter a single integer.")

print("Enter roads:")

# Union for each road
for _ in range(m):
    while True:
        try:
            a, b = map(int, input().split())
            break
        except ValueError:
            print("Invalid input. Please enter two integers separated by a space.")
    dsu.union(a, b)

# Read number of queries
while True:
    try:
        q = int(input("Enter number of queries (a single integer): "))
        break
    except ValueError:
        print("Invalid input. Please enter a single integer.")

print("Enter queries:")

# Process queries
for _ in range(q):
    while True:
        try:
            x, y = map(int, input().split())
            break
        except ValueError:
            print("Invalid input. Please enter two integers separated by a space.")
    if dsu.find(x) == dsu.find(y):
        print("YES")
    else:
        print("NO")
```
