```python
# Segment Tree for Min and Max

class SegmentTree:
    def __init__(self, arr):
        self.n = len(arr)
        self.minTree = [0] * (4 * self.n)
        self.maxTree = [0] * (4 * self.n)
        self.build(1, 0, self.n - 1, arr)

    def build(self, node, start, end, arr):
        if start == end:
            self.minTree[node] = arr[start]
            self.maxTree[node] = arr[start]
        else:
            mid = (start + end) // 2
            self.build(2 * node, start, mid, arr)
            self.build(2 * node + 1, mid + 1, end, arr)

            self.minTree[node] = min(
                self.minTree[2 * node],
                self.minTree[2 * node + 1]
            )

            self.maxTree[node] = max(
                self.maxTree[2 * node],
                self.maxTree[2 * node + 1]
            )

    def rangeMin(self, node, start, end, l, r):
        if r < start or end < l:
            return float('inf')

        if l <= start and end <= r:
            return self.minTree[node]

        mid = (start + end) // 2

        left = self.rangeMin(2 * node, start, mid, l, r)
        right = self.rangeMin(2 * node + 1, mid + 1, end, l, r)

        return min(left, right)

    def rangeMax(self, node, start, end, l, r):
        if r < start or end < l:
            return -float('inf')

        if l <= start and end <= r:
            return self.maxTree[node]

        mid = (start + end) // 2

        left = self.rangeMax(2 * node, start, mid, l, r)
        right = self.rangeMax(2 * node + 1, mid + 1, end, l, r)

        return max(left, right)

    def update(self, node, start, end, idx, val):
        if start == end:
            self.minTree[node] = val
            self.maxTree[node] = val
        else:
            mid = (start + end) // 2

            if idx <= mid:
                self.update(2 * node, start, mid, idx, val)
            else:
                self.update(2 * node + 1, mid + 1, end, idx, val)

            self.minTree[node] = min(
                self.minTree[2 * node],
                self.minTree[2 * node + 1]
            )

            self.maxTree[node] = max(
                self.maxTree[2 * node],
                self.maxTree[2 * node + 1]
            )


# -------- MAIN PROGRAM --------

n = int(input("Enter number of sensors: "))
arr = list(map(int, input("Enter temperature readings: ").split()))

seg = SegmentTree(arr)

q = int(input("Enter number of queries: "))

print("Enter queries:")

for _ in range(q):

    query = input().split()

    if query[0] == "RangeMin":
        l = int(query[1])
        r = int(query[2])

        ans = seg.rangeMin(1, 0, n - 1, l, r)
        print(ans)

    elif query[0] == "RangeMax":
        l = int(query[1])
        r = int(query[2])

        ans = seg.rangeMax(1, 0, n - 1, l, r)
        print(ans)

    elif query[0] == "Update":
        i = int(query[1])
        x = int(query[2])

        seg.update(1, 0, n - 1, i, x)
```
